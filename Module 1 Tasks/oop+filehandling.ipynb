```python
class Account:

    def __init__(self, account_number, balance):
        self.account_number = account_number
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print("Amount deposited:", amount)

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print("Amount withdrawn:", amount)
        else:
            print("Insufficient balance")

    def display_balance(self):
        print("Account Number:", self.account_number)
        print("Balance:", self.balance)


class SavingsAccount(Account):

    def __init__(self, account_number, balance, interest_rate):
        super().__init__(account_number, balance)
        self.interest_rate = interest_rate

    def add_interest(self):
        interest = self.balance * self.interest_rate / 100
        self.balance += interest
        print("Interest added:", interest)


# Create object
s1 = SavingsAccount("12345", 10000, 5)

s1.deposit(2000)
s1.withdraw(3000)
s1.add_interest()
s1.display_balance()
```

```python
class Account:

    def __init__(self, account_number, balance):
        self.account_number = account_number
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print("Deposited:", amount)

    def display_balance(self):
        print("Account Number:", self.account_number)
        print("Balance:", self.balance)


# Savings Account Class
class SavingsAccount(Account):

    MIN_BALANCE = 500

    def withdraw(self, amount):

        if self.balance - amount >= self.MIN_BALANCE:
            self.balance -= amount
            print("Withdrawn:", amount)
        else:
            print("Minimum balance must be maintained (500)")


# Current Account Class
class CurrentAccount(Account):

    OVERDRAFT_LIMIT = -5000

    def withdraw(self, amount):

        if self.balance - amount >= self.OVERDRAFT_LIMIT:
            self.balance -= amount
            print("Withdrawn:", amount)
        else:
            print("Overdraft limit exceeded!")


# Testing SavingsAccount
print("\nSavings Account Example")
s1 = SavingsAccount("S123", 2000)

s1.deposit(500)
s1.withdraw(1800)
s1.display_balance()


# Testing CurrentAccount
print("\nCurrent Account Example")
c1 = CurrentAccount("C456", 2000)

c1.withdraw(6000)
c1.display_balance()
```

```python
import json


# Step 1: Create Student class
class Student:

    def __init__(self, name, age, marks):
        self.name = name
        self.age = age
        self.marks = marks


# Step 2: Create student objects
s1 = Student("Alice", 20, 85)
s2 = Student("Bob", 21, 90)

students = [s1, s2]


# Step 3: Convert student objects into dictionary format
student_data = []

for s in students:
    student_data.append({
        "name": s.name,
        "age": s.age,
        "marks": s.marks
    })


# Step 4: Save dictionary data into JSON file
with open("students.json", "w") as file:
    json.dump(student_data, file, indent=4)

print("Students saved successfully!")


# Step 5: Read data from JSON file
loaded_students = []

with open("students.json", "r") as file:
    data = json.load(file)


# Step 6: Convert dictionary data back into Student objects
for item in data:
    loaded_students.append(
        Student(item["name"], item["age"], item["marks"])
    )


# Step 7: Display loaded students
print("\nLoaded Students:")

for s in loaded_students:
    print(s.name, s.age, s.marks)
```

```python
import json


# Step 1: Create Student class
class Student:

    def __init__(self, name, age, marks):
        self.name = name
        self.age = age
        self.marks = marks


# Step 2: Take number of students from user
n = int(input("Enter number of students: "))

students = []


# Step 3: Take student details from user
for i in range(n):

    print(f"\nEnter details of student {i+1}")

    name = input("Enter name: ")
    age = int(input("Enter age: "))
    marks = int(input("Enter marks: "))

    students.append(Student(name, age, marks))


# Step 4: Convert student objects into dictionary format
student_data = []

for s in students:
    student_data.append({
        "name": s.name,
        "age": s.age,
        "marks": s.marks
    })


# Step 5: Save into JSON file
with open("students.json", "w") as file:
    json.dump(student_data, file, indent=4)

print("\nStudents saved successfully!")


# Step 6: Read data back from JSON file
loaded_students = []

with open("students.json", "r") as file:
    data = json.load(file)


# Step 7: Convert dictionary data back into Student objects
for item in data:
    loaded_students.append(
        Student(item["name"], item["age"], item["marks"])
    )


# Step 8: Display loaded student details
print("\nLoaded Students:")

for s in loaded_students:
    print(s.name, s.age, s.marks)
```

```python
# student management system with CRUD operations
import json 
class Student:
    def __init__(self, name, age, marks):
        self.name = name
        self.age = age
        self.marks = marks

    def display(self):
        print(f"Name: {self.name}, Age: {self.age}, Marks: {self.marks}")
class StudentManagementSystem:  
    def __init__(self):
        self.students = []
        self.load_students()

    def load_students(self):
        try:
            with open("students.json", "r") as file:
                data = json.load(file)
                for item in data:
                    self.students.append(Student(item["name"], item["age"], item["marks"]))
        except FileNotFoundError:
            pass

    def save_students(self):
        student_data = [{"name": s.name, "age": s.age, "marks": s.marks} for s in self.students]
        with open("students.json", "w") as file:
            json.dump(student_data, file, indent=4)

    def add_student(self):
        name = input("Enter name: ")
        age = int(input("Enter age: "))
        marks = int(input("Enter marks: "))
        self.students.append(Student(name, age, marks))
        self.save_students()
        print("Student added successfully!")

    def display_students(self):
        if not self.students:
            print("No students found.")
            return
        for student in self.students:
            student.display()

    def update_student(self):
        name = input("Enter name of student to update: ")
        for student in self.students:
            if student.name == name:
                student.age = int(input("Enter new age: "))
                student.marks = int(input("Enter new marks: "))
                self.save_students()
                print("Student updated successfully!")
                return
        print("Student not found.")

    def delete_student(self):
        name = input("Enter name of student to delete: ")
        for i, student in enumerate(self.students):
            if student.name == name:
                del self.students[i]
                self.save_students()
                print("Student deleted successfully!")
                return
        print("Student not found.")
def main():
    system = StudentManagementSystem()
    while True:
        print("\n1. Add Student")
        print("2. Display Students")
        print("3. Update Student")
        print("4. Delete Student")
        print("5. Exit")
        choice = input("Enter your choice: ")
        if choice == "1":
            system.add_student()
        elif choice == "2":
            system.display_students()
        elif choice == "3":
            system.update_student()
        elif choice == "4":
            system.delete_student()
        elif choice == "5":
            break
        else:
            print("Invalid choice. Please try again.")
            
if __name__ == "__main__":
    main()  
    
```
