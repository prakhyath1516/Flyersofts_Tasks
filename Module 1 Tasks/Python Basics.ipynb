```python
#calculator program with error handling for division by zero and invalid operators

def calculator(num1, num2, operator):
    if operator == "+":
        return num1 + num2
    elif operator == "-":
        return num1 - num2
    elif operator == "*":
        return num1 * num2
    elif operator == "/":
        if num2 != 0:
            return num1 / num2
        else:
            return "Error: Cannot divide by zero"
    elif operator == "%":
        if num2 != 0:
            return num1 % num2
        else:
            return "Error: Cannot perform modulus with zero"
    elif operator == "**":
        return num1 ** num2
    else:
        return "Error: Invalid operator"

# Example usage
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
operator = input("Enter operator (+, -, *, /, %, **): ")

result = calculator(num1, num2, operator)
print(f"{num1} {operator} {num2} = {result}")
```

```python
#Number Classification Program
def classify_number(num):
    if num > 0:
        return "Positive"
    elif num < 0:
        return "Negative"
    else:
        return "Zero"
# Example usage
num = float(input("Enter a number to classify: "))
classification = classify_number(num)
print(f"The number {num} is {classification}.")
```

```python
#Pattern Printing Program
def print_pattern(n):
    for i in range(1, n + 1):
        print("* " * i)
# Example usage
n = int(input("Enter the number of rows for the pattern: "))
print_pattern(n)    
```

```python
#build a guessing game and add attempt counter and scoring system
import random 
def guessing_game():
    number_to_guess = random.randint(1, 100)
    attempts = 0
    score = 100
    
    print("Welcome to the Guessing Game!")
    print("I have selected a number between 1 and 100. Can you guess it?")
    
    while True:
        try:
            guess = int(input("Enter your guess: "))
            attempts += 1
            
            if guess < number_to_guess:
                print("Too low! Try again.")
                score -= 10
            elif guess > number_to_guess:
                print("Too high! Try again.")
                score -= 10
            else:
                print(f"Congratulations! You've guessed the number {number_to_guess} in {attempts} attempts.")
                print(f"Your final score is: {score}")
                break
        except ValueError:
            print("Invalid input. Please enter a valid integer.")
# Start the game
guessing_game()
```
